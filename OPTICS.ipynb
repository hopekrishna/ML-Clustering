```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import OPTICS
import pandas as pd
from sklearn.datasets import make_blobs
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
x
```

```python
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)
```

```python
# Apply OPTICS clustering
# Initialize the OPTICS model with min_samples=5 and the maximum distance (eps) for points to be considered neighbors
optics = OPTICS(min_samples=5, eps=0.05, min_cluster_size=0.1)
optics.fit(X)
```

```python
# Get the labels and plot the clustering results
labels = optics.labels_
```

```python
# Step 4: Create the plot
plt.figure(figsize=(8, 6))

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o')
# Plot noise points as gray
plt.scatter(X[labels == -1, 0], X[labels == -1, 1], color='gray', marker='x', label='Noise')

plt.title("OPTICS Clustering")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
plt.show()
```
