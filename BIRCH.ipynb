```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import Birch
import pandas as pd
from sklearn.datasets import make_blobs
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
x=dataset.iloc[:,[3,4]].values
```

```python
x
```

```python
X, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.60, random_state=0)
```

```python

# Initialize the BIRCH model
birch = Birch(n_clusters=3)  # Set the number of clusters you want to find
```

```python
# Fit the BIRCH model to the data
birch.fit(X)
```

```python
# Get the cluster labels for each point
labels = birch.labels_
```

```python
# Visualize the result
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', s=50)
plt.title('BIRCH Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
# Print cluster centroids (BIRCH doesn't directly provide centroids like k-means)
# But you can still get the cluster centers from the final clustering
print("Cluster Centers (approximate):")
for i in range(birch.n_clusters):
    cluster_points = X[labels == i]
    center = np.mean(cluster_points, axis=0)
    print(f"Cluster {i}: Center = {center}")
```
